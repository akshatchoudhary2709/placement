```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
data.drop(['sl_no'],axis=1,inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.columns.unique()
```

```python
data.drop(['specialisation'],axis=1,inplace=True)
```

```python
data.drop(['mba_p'],axis=1,inplace=True)
```

```python
data.columns
```

```python
lst=['gender', 'ssc_b', 'hsc_b', 'hsc_s',
       'degree_t', 'workex']

for i in lst:
    print(i,': ',data[i].unique())
```

```python
gender=['M','F']
for i in gender:
    if i=="M":
        data['gender']=data['gender'].replace('M',1)
    else:
        data['gender']=data['gender'].replace('F',0)
        
data['gender']   
```

```python
board_ssc=pd.get_dummies(data[['ssc_b']],drop_first=True)
```

```python
board_ssc
```

```python
board_hsc=pd.get_dummies(data[['hsc_b']],drop_first=True)
```

```python
board_hsc
```

```python
data.head()
```

```python
data.head()
```

```python
stream=data[['hsc_s']]
stream=pd.get_dummies(stream,drop_first=True)
```

```python
stream
```

```python
stream2=data[['degree_t']]
stream2=pd.get_dummies(stream2,drop_first=True)
stream2
```

```python
workex=data[['workex']]
workex=pd.get_dummies(workex,drop_first=True)
```

```python
data=pd.concat([data,board_ssc,board_hsc,stream,stream2,workex],axis=1)
```

```python
data.head()
```

```python
data.drop(['ssc_b','hsc_b','degree_t','workex','hsc_s'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data=data[['gender', 'ssc_p', 'hsc_p', 'degree_p', 'etest_p',
       'salary', 'ssc_b_Others', 'hsc_b_Others', 'hsc_s_Commerce',
       'hsc_s_Science', 'degree_t_Others', 'degree_t_Sci&Tech', 'workex_Yes', 'status']]
```

```python
data.head()
```

```python
data['status']=data['status'].map({'Placed':1,'Not Placed':0})
```

```python
data['status'].head()
```

```python
data.head()
```

```python
data.drop('salary',axis=1,inplace=True)
```

```python
plt.figure(figsize=(12,16))
sns.heatmap(data.corr(),annot=True)
```

```python
X=data.iloc[:,:12]
```

```python
X.head()
```

```python
X.shape
```

```python
y=data['status']
```

```python
y.head()
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=27)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
lo_reg=LogisticRegression()
```

```python
lo_reg.fit(X_train,y_train)
```

```python
y_pred=lo_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,r2_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
